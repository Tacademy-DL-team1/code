중복 방지 및 체결 코드 구현 후 테스트

```python
import cv2
import numpy as np
# from ultralytics import RTDETR  # 모델 로드 필요 시 주석 해제

def shrink_hook_box(box, keep_ratio=0.3):
    """
    후크 박스의 상단 일정 비율(기본 30%)만 남기고 아래를 잘라냅니다.
    밧줄을 제외한 '금속 고리' 부분만 추출하여 정확한 겹침 계산을 하기 위함입니다.
    """
    xmin, ymin, xmax, ymax = box
    height = ymax - ymin
    new_ymax = ymin + (height * keep_ratio)
    return [int(xmin), int(ymin), int(xmax), int(new_ymax)]

def check_is_touching(box_hook, box_line):
    """
    [핵심 수정] 비율을 계산하지 않고, 두 박스가 단 1픽셀이라도 겹치는지 여부만 판별합니다.
    파이프(Lifeline)가 아무리 얇아도 후크 상단과 만나기만 하면 체결로 인정됩니다.
    """
    x_left = max(box_hook[0], box_line[0])
    y_top = max(box_hook[1], box_line[1])
    x_right = min(box_hook[2], box_line[2])
    y_bottom = min(box_hook[3], box_line[3])

    # 교차 영역의 가로나 세로가 0 이하면 전혀 겹치지 않는 상태입니다.
    if x_right <= x_left or y_bottom <= y_top:
        return False

    # 겹치는 영역(교집합)이 존재하므로 True 반환
    return True

# ---------------------------------------------------------
# 1. 모델 예측 환경 설정
# ---------------------------------------------------------
image_path = "/content/미체결_철근.jpg"  # 테스트할 이미지 경로 입력
img_cv = cv2.imread(image_path)
h, w, _ = img_cv.shape

# model = RTDETR("/content/runs/detect/RTDETR/hook_detection-2/weights/best.pt")

# 박스 중복 제거를 위해 conf=0.50, iou=0.40 적용
results = model.predict(
    source=image_path,
    conf=0.40,
    iou=0.40,
    device=0
)

boxes = results[0].boxes
names = model.names

# 클래스별 리스트 초기화
hooks = []
lifelines = []
workers = []
harnesses = []

# ---------------------------------------------------------
# 2. 예측 결과 분류 (파싱)
# ---------------------------------------------------------
for box in boxes:
    coords = [int(c) for c in box.xyxy[0].tolist()]
    cls_id = int(box.cls[0])
    cls_name = names[cls_id]
    conf_score = float(box.conf[0])

    if cls_name == 'hook':
        hooks.append({'box': coords, 'conf': conf_score, 'status': 'Detached'})
    elif cls_name == 'lifeline':
        lifelines.append({'box': coords})
    elif cls_name == 'worker':
        workers.append({'box': coords})
    elif cls_name == 'harness':
        harnesses.append({'box': coords})

# ---------------------------------------------------------
# 3. 체결 유무 판별 (단순 접촉 판별 + 라이프라인 좌우 확장)
# ---------------------------------------------------------
LIFELINE_PADDING = 150  # 모델이 파이프를 짧게 끊어서 인식할 때를 대비한 좌우 가상 확장 길이

for hook in hooks:
    is_attached = False  # 기본 상태는 미체결(Detached)

    # 밧줄을 뺀 '상단 30% 금속 고리' 영역만 임시 추출
    shrunk_hook_box = shrink_hook_box(hook['box'], keep_ratio=0.3)

    for line in lifelines:
        padded_line = line['box'].copy()

        # 파이프 박스의 좌우 범위를 늘려줍니다. (이미지 경계 안쪽으로 제한)
        padded_line[0] = max(0, padded_line[0] - LIFELINE_PADDING)
        padded_line[2] = min(w, padded_line[2] + LIFELINE_PADDING)

        # 늘어난 파이프 선과 후크 고리가 조금이라도 교차하는지 체크
        if check_is_touching(shrunk_hook_box, padded_line):
            is_attached = True
            break  # 하나라도 닿은 파이프가 있다면 이 후크는 체결된 것이므로 루프 종료

    if is_attached:
        hook['status'] = 'Attached'
    else:
        hook['status'] = 'Detached'

# ---------------------------------------------------------
# 4. 시각화 및 결과 드로잉
# ---------------------------------------------------------

# Worker (작업자): 파란색
for worker in workers:
    x1, y1, x2, y2 = worker['box']
    cv2.rectangle(img_cv, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(img_cv, "Worker", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

# Harness (하네스): 보라색
for harness in harnesses:
    x1, y1, x2, y2 = harness['box']
    cv2.rectangle(img_cv, (x1, y1), (x2, y2), (255, 0, 255), 2)
    cv2.putText(img_cv, "Harness", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 255), 2)

# Lifeline (생명줄 파이프): 노란색
for line in lifelines:
    x1, y1, x2, y2 = line['box']
    cv2.rectangle(img_cv, (x1, y1), (x2, y2), (0, 255, 255), 2)
    cv2.putText(img_cv, "Lifeline", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)

# Hook (후크): 체결은 초록색, 미체결은 빨간색
for hook in hooks:
    x1, y1, x2, y2 = hook['box']  # 그릴 때는 원본 전체 박스를 사용합니다.

    if hook['status'] == 'Attached':
        color = (0, 255, 0)  # 초록색
        label = f"Hook (Attached) {hook['conf']:.2f}"
    else:
        color = (0, 0, 255)  # 빨간색
        label = f"Hook (Detached) {hook['conf']:.2f}"

    cv2.rectangle(img_cv, (x1, y1), (x2, y2), color, 3)
    cv2.putText(img_cv, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

# ---------------------------------------------------------
# 5. 결과 파일 저장
# ---------------------------------------------------------
output_path = "result_visualized_fixed4.jpg"
cv2.imwrite(output_path, img_cv)
print(f"새로운 판별 로직이 적용된 결과 이미지가 {output_path}에 저장되었습니다.")
```
